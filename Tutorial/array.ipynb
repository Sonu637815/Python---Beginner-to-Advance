```python
# append() : to add element to the end of the list
# extend() : to extend all elements of a list to the another list
# insert() : to insert an element at the another index
# remove() : to remove an element from the list
# pop() : to remove elements return element at the given index
# clear() : to remove all elements from the list
# index() : to return the index of the first matched element
# count() : to count of number of elements passed as an argument
# sort() : to sort the elements in ascending order by default
# reverse() : to reverse order element in a list
# copy() : to return a copy of elements in a list
```

```python
# Defining and declaring an array
arr = [10, 20, 30, 40, 50]
print(arr)

# Accessing array elements 
print(arr[0])
print(arr[1])
print(arr[2])
print(arr[-2])
print(arr[-1])
```

```python
brands = ["Apple", "Samsung", "Oppo", "Vivo", "Oneplus"]
print(brands)

# length of the array 
print(len(brands))

# adding element to the array
brands.append("Nothing")
print(brands)
```

```python
# removing elements from the array
colors = ["violet", "indigo","blue", "green", "yellow", "orange", "red"]
print(colors)

del colors[2]
print(colors)

colors.remove("indigo")
print(colors)

colors.pop(4)
print(colors)
```

```python
# Modifying elements of an array
fruits = ["apple", "banana", "mango", "grapes", "orange"]
print(fruits)

fruits[2] = "kiwi"
print(fruits)

fruits[-1] = "watermelon"
print(fruits)
```

```python
# concatenating two arrays
arr1 = [1, 2, 3]
print(arr1)

arr1 = arr1 + [4, 5, 6]
print(arr1)
```

```python
# repeating elements in an array
repeat = ["s"]
print(repeat)

repeat = repeat * 6
print(repeat)
```

```python
# slicing arrays
fruits = ["apple", "banana", "mango", "grapes", "orange"]
print(fruits)
print(fruits[1:4])
print(fruits[ : 3])
print(fruits[-4:])
print(fruits[-3:-1])
```

```python
# Declaring and defining f=g multidimensional array
multd = [[1,2], [3,4], [5,6], [7,8]]
print(multd)

print(multd[0])
print(multd[1])
print(multd[2][1])
print(multd[3][0])
```
